```python
def get_date(dt):
    try:
        return datetime.strptime(dt, "%Y-%m-%d %H:%M:%S+00:00")
    except:
        try:
            return datetime.strptime(dt, "%Y-%m-%d %H:%M:%S")
        except:
            return None
```

```python
zips = pandas.read_csv('zip_code_database.csv')
zips.columns
city_dict = {}
for city, lat,lon in zips[['acceptable_cities', 'latitude','longitude']].itertuples(index=False):
    if str(city) is not "nan":
        for ci in str(city).split(','):
            city_dict[ci] = (lat, lon)       
```

```python
def generate_arc_2013(new_data):
    arcs = []
    multiple = int(255/len(new_data))
    for x in range(len(new_data)-1):
        if x+1 != len(new_data):
            arcs.append({
                "origin":{
                    "latitude":new_data[x][2][0],
                    "longitude":new_data[x][2][1]
                },
                "destination":{
                    "latitude":new_data[x+1][2][0],
                    "longitude":new_data[x+1][2][1]
                },
                "options":{
                    "strokeColor":'rgba(%i, 0, 255, 0.1)' % (int(x*multiple))
                }
            })
    return arcs

def generate_bubble_2013(new_data, name):
    smonday, sday, (slat, slon) = new_data[0]
    emonday, eday, (elat, elon) = new_data[-1]
    
    return [{
    "borderWidth":0,
    "name":"Start of " + name,
    "radius":"5",
    "fillOpacity": 0.25,
    "fillKey":"start_2013",
    "latitude": slat,
    "longitude": slon
      },{
    "borderWidth":0,
    "fillOpacity": 0.25,
    "name":"End of " + name,
    "radius":"5",
    "fillKey":"end_2013",
    "latitude": elat,
    "longitude": elon
      }]


def generate_arc_2014(new_data):
    arcs = []
    multiple = int(255/len(new_data))
    for x in range(len(new_data)-1):
        if x+1 != len(new_data):
            arcs.append({
                "origin":{
                    "latitude":new_data[x][2][0],
                    "longitude":new_data[x][2][1]
                },
                "destination":{
                    "latitude":new_data[x+1][2][0],
                    "longitude":new_data[x+1][2][1]
                },
                "options":{
                    "strokeColor":'rgba(%i, 0, 255, 0.1)' % (int(x*multiple))
                }
            })
    return arcs

def generate_bubble_2014(new_data, name):
    smonday, sday, (slat, slon) = new_data[0]
    emonday, eday, (elat, elon) = new_data[-1]
    
    return [{
    "borderWidth":0,
    "name":"Start of " + name,
    "radius":"5",
    "fillOpacity": 0.25,
    "fillKey":"start_2013",
    "latitude": slat,
    "longitude": slon
      },{
    "borderWidth":0,
    "fillOpacity": 0.25,
    "name":"End of " + name,
    "radius":"5",
    "fillKey":"end_2013",
    "latitude": elat,
    "longitude": elon
      }]
```
